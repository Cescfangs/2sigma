```python
train, origin_features = twosigmafunc.preprocess()
```

```python
train.head()
```

```python
twosigmafunc.add_nans(train, origin_features)
```

```python
train = train.fillna(train.median())
```

```python
train.head()
```

```python
twosigmafunc.add_diffs(train, origin_features)
```

```python
train.head()
```

```python
train['tec20-30'] = train.technical_20 - train.technical_30
```

# Get train and test data 

```python
X_train, y_train, X_test, y_test = twosigmafunc.split_data(train, twosigmafunc.origin_features(train))
```

```python
lr_features = ['tec20-30', 'technical_20_diff']
```

```python
X_train.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.linear_model import LinearRegression
```

```python
etr0 = ExtraTreesRegressor(n_estimators=128, n_jobs=-1, max_depth=6, max_features=0.5, random_state=1)
```

```python
etr0.fit(X_train, y_train)
y_etr = etr0.predict(X_test)
```

```python
twosigmafunc.R_score(y_etr, y_test)
```

```python
lr = LinearRegression(n_jobs=-1)
lr.fit(X_train[lr_features], y_train)
```

```python
X_train['y_lr'] = lr.predict(X_train[lr_features])
X_test['y_lr'] = lr.predict(X_test[lr_features])
```

```python
etr1 = ExtraTreesRegressor(n_estimators=128, n_jobs=-1, max_depth=6, max_features=0.5, random_state=1)
etr1.fit(X_train, y_train)
y_etr1 = etr1.predict(X_test)
twosigmafunc.R_score(y_etr1, y_test)
```

```python
imp = pd.DataFrame({'feature':X_test.columns, 'important':etr1.feature_importances_}).sort_values('important', ascending=0)
```
