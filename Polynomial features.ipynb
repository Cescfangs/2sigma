```python
import numpy as np
import pandas as pd
import twosigmafunc
```

```python
train = pd.read_hdf('train.h5')
```

```python
train.fillna(train.median(), inplace=True)
```

```python
poly = 2
```

```python
excl = ['id', 'timestamp', 'y']
origin_features = [col for col in train.columns if col not in excl]
poly_features = [col + '_p' for col in origin_features]
```

```python
for col in origin_features:
    train[col + '_p'] = train[col]**poly
```

```python
corrs = {}
for col in poly_features:
    corrs[col] = np.corrcoef(train[col], train.y)[0][1]
```

```python
corrs = sorted(corrs.items(), key=lambda x : np.abs(x[1]), reverse=True)
```

```python
corrs[:10]
```

```python
X_train, y_train, X_test, y_test = twosigmafunc.split_data(train, origin_features + poly_features)
```

```python
del train
```

```python
X_test.info()
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from collections import OrderedDict
```

```python
low_y_cut = -0.075
high_y_cut = 0.075
y_is_above_cut = (y_train > high_y_cut)
y_is_below_cut = (y_train < low_y_cut)
y_is_within_cut = (~y_is_above_cut & ~y_is_below_cut)
```

```python
lr = LinearRegression(n_jobs=-1)
poly_score = {}
for feature in poly_features:
    lr.fit(X_train.loc[y_is_within_cut, feature].reshape(-1,1), y_train.loc[y_is_within_cut])
    y_pred = lr.predict(X_test[feature].reshape(-1,1)).clip(low_y_cut, high_y_cut)
    poly_score[feature] = twosigmafunc.R_score(y_pred, y_test)
```

```python
poly_score = sorted(poly_score.items(),key=lambda x:x[1], reverse=True)
```

```python
poly_score[:5]
```

```python
lr.fit(X_train.loc[y_is_within_cut, ['technical_20_p', 'technical_30_p']], y_train.loc[y_is_within_cut])
```

```python
y_pred = lr.predict(X_test[['technical_20_p', 'technical_30_p']])
score = twosigmafunc.R_score(y_pred, y_test)
```

```python
score
```

```python
lr = LinearRegression(n_jobs=-1)
lr.fit(X_train.loc[y_is_within_cut, ['technical_20_p', 'technical_30_p','technical_20', 'technical_30']], y_train.loc[y_is_within_cut])
```

```python
y_pred = lr.predict(X_test[['technical_20_p', 'technical_30_p','technical_20', 'technical_30']])
score = twosigmafunc.R_score(y_pred, y_test)
score
```

```python
lr = LinearRegression(n_jobs=-1, normalize=True)
lr.fit(X_train.loc[y_is_within_cut, ['technical_20_p', 'technical_30_p','technical_20', 'technical_30']], y_train.loc[y_is_within_cut])
```

```python
y_pred = lr.predict(X_test[['technical_20_p', 'technical_30_p','technical_20', 'technical_30']])
score = twosigmafunc.R_score(y_pred, y_test)
score
```

```python
ridge = Ridge(normalize=True)
ridge.fit(X_train.loc[y_is_within_cut, ['technical_20_p', 'technical_30_p','technical_20', 'technical_30']], y_train.loc[y_is_within_cut])
```

```python
y_pred = ridge.predict(X_test[['technical_20_p', 'technical_30_p','technical_20', 'technical_30']])
score = twosigmafunc.R_score(y_pred, y_test)
score
```
