```python
import numpy as np
def fill_logic(x, col, mean):
    if x == x:
        return x
    elif np.abs(col) < 1e-3:
        return mean
    else:
        return 0.0
```

```python
def fill_tech20(data, cols):
    friends = data[cols]
    col_max = friends.max()
    tech20_mean = data['technical_29'].mean()
    frid_mean = data.mean()
    for col in cols:
        data[col] = map(fill_logic(col=train.technical_20, mean=frid_mean[col]))
```
