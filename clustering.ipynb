```python
train_sm = train[['id', 'technical_20', 'technical_30', 'technical_13']]
```

```python
train_sm.head()
```

```python
train_sm.isnull().sum()
```

```python
train_sm = train_sm.loc[train.technical_20.notnull()]
```

```python
sm_med = train_sm.groupby('id').mean()
```

```python
sm_med.fillna(sm_med.median(), inplace=True)
```

```python
sm_med.index
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
inert = []
for n in range(3,10):
    km = KMeans(n_clusters=n, n_jobs=-2, random_state=11, n_init=20)
    km.fit(sm_med[['technical_20', 'technical_30', 'technical_13']])
    score = silhouette_score(sm_med[['technical_20', 'technical_30', 'technical_13']], km.labels_)
    inert.append((n,score))
```

```python
inert
```

```python
km = KMeans(n_clusters=5, n_jobs=-2, random_state=11, n_init=30)
km.fit(sm_med[['technical_20', 'technical_30', 'technical_13']])
```

```python
km.labels_
```

```python
sm_med.index
```

```python
from collections import defaultdict
group_id = defaultdict(list)
for id_, group in zip(sm_med.index, km.labels_):
    group_id[group].append(id_)
```

```python
group_id
```

```python
train.id.unique().shape
```

```python
group_id[5]
```

```python
def label(id):
    if id in group_id[0]:
        return 'a'
    elif id in group_id[1]:
        return 'b'
    elif id in group_id[2]:
        return 'c'
    elif id in group_id[3]:
        return 'd'
    elif id in group_id[4]:
        return 'e'
    elif id in group_id[5]:
        return 'f'
    elif id in group_id[6]:
        return 'g'
    elif id in group_id[7]:
        return 'h'
    elif id in group_id[8]:
        return 'i'
    else:
        return 'n'

    
```

```python
d_mean= train.median(axis=0)
n = train.isnull().sum(axis=1)
for c in train.columns:
    train[c + '_nan_'] = pd.isnull(train[c])
    d_mean[c + '_nan_'] = 0
train = train.fillna(d_mean)
train['znull'] = n
```
